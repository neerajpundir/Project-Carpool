```python
from tkinter import *
from PIL import Image, ImageTk
```

```python
# Screen
scr = Tk()

# Main Frame 
def main_frame():
    scr.state("zoomed")
    scr.configure(bg="powder blue")
    lbl_title = Label(scr, text= "Car Pooling", font=('',80,"bold","underline"), bg= "powder blue")
    lbl_title.pack()

    logo = Image.open('img\logo.jpg').resize((400,150))
    logotk = ImageTk.PhotoImage(logo, master =scr )
    lbl_logo = Label(scr, image= logotk)
    lbl_logo.place(relx=0,rely=0)

    logo2 = Image.open('img\carpool.jpg').resize((400,150))
    logotk2 = ImageTk.PhotoImage(logo2, master = scr )
    lbl_logo2 = Label(scr, image= logotk2)
    lbl_logo2.place(relx=0.76,rely=0)

    scr.mainloop()
```

```python
# Login frame
def login_frame():
    
    # Adding a login frame
    login_frm = Frame(scr, highlightbackground="black" , highlightthickness=3)
    login_frm.place(x=400, y=250, relwidth =.5, relheight=.45)
    
    # Adding Fields Username and Password
    lbl_user = Label(login_frm, text="Username:",font=('',20,"bold"))
    lbl_pass = Label(login_frm, text="Password:",font=('',20,"bold"))
    e_user = Entry(login_frm, font=('',20,"bold"), bd=3 )
    e_user.focus()
    e_pass = Entry(login_frm, font=('',20,"bold"), bd=3, show = "*" )
    lbl_user.place(x=120,y=50)
    e_user.place(x=300,y=50)
    lbl_pass.place(x=120,y=110)
    e_pass.place(x=300,y=110)
    
    # Funtion for login user
    def login():
        login_frm.destroy()
        index_frame()
        
    # Adding Button Login, Reset
    btn_login = Button(login_frm, text ="Login",font=('',18,"bold"), bd=3, command=login,width=15)
    btn_login.place(x=120,y=180)
    btn_reset = Button(login_frm, text ="Reset",font=('',18,"bold"),bd=3,width=15)
    btn_reset.place(x=420,y=180)
   
    # Funtion for calling New user Frame and Destroying Previous frame
    def newuser():
        login_frm.destroy()
        newuser_frame()
    
    # Add New user button 
    btn_new_user = Button(login_frm, text ="Create a New Account",font=('',16,"bold"), bd=3, command= newuser, width=30)
    btn_new_user.place(x=200,y=250)
```

```python
# New User Frame    
def newuser_frame():

    newuser_frm = Frame(scr, highlightbackground="black" , highlightthickness=3)
    newuser_frm.place(x=400, y=250, relwidth =.5, relheight=.5)

    #Username Textfield    
    lbl_uname = Label(newuser_frm, text="Username:",font=('',20,"bold"))
    lbl_uname.place(x = 130,y = 50)
    e_username = Entry(newuser_frm,font=('',20,"bold"), bd=3)
    e_username.place(x=320,y=50)

    # Password Textfield
    lbl_pass = Label(newuser_frm, text="Password:",font=('',20,"bold"))
    lbl_pass.place(x=130,y=100)
    e_password = Entry(newuser_frm, font=('',20,'bold'), bd=3)
    e_password.place(x=320, y=100)

    # Email textfield
    lbl_email = Label(newuser_frm, text="Email:",font=('',20,"bold"))
    lbl_email.place(x=130,y=150)
    e_email = Entry(newuser_frm, font=('',20,'bold'), bd=3)
    e_email.place(x=320, y=150)

    # Mobile Number textfield
    lbl_number = Label(newuser_frm, text="Mobile No.:",font=('',20,"bold"))
    lbl_number.place(x=130,y=200)
    e_number = Entry(newuser_frm, font=('',20,'bold'), bd=3)
    e_number.place(x=320, y=200)

    # Button for Adding New User
    btn_reg = Button(newuser_frm,text="Register",font=('',20,"bold"),bd=3)
    btn_reg.place(x=150,y=250)

    # Funtion for return to Login Frame
    def back_home():
        newuser_frm.destroy()
        login_frame()

    # Back Button 
    btn_back = Button(newuser_frm,text="Back",font=('',20,"bold"),bd=3, command=back_home)
    btn_back.place(x=500,y=250)
```

```python
def index_frame():
    index_frm = Frame(scr, highlightbackground="black" , highlightthickness=3)
    index_frm.place(x=400, y=250, relwidth =.45, relheight=.45)
  
    lbl_user = Label(index_frm,text="Welcome, User", font= ('',12,"bold"),fg="green")
    lbl_user.place(relx=0,rely=0)

    # Funtion for return to Login Frame
    def back_home():
        index_frm.destroy()
        login_frame()
  
    btn_logout = Button(index_frm,text="Logout",font=('',12,"bold"),bd=3, command=back_home)
    btn_logout.place(relx=.89,rely=0)
  
    # Funtion for return to Login Frame
    def search():
        index_frm.destroy()
        serach_pool()
     
    def newpool():
        index_frm.destroy()
        new_pool()   
  
    btn_search = Button(index_frm,text="Search CarPool",font=('',16,"bold"),bd=3,width=15, command=search)
    btn_search.place(relx=.35,rely=.2)
  
    btn_newpool = Button(index_frm,text="Create CarPool",font=('',16,"bold"),bd=3,width=15, command=newpool)
    btn_newpool.place(relx=.35,rely=.4)
  
    btn_profile = Button(index_frm,text="User Profile",font=('',16,"bold"),bd=3,width=15)
    btn_profile.place(relx=.35,rely=.6)
```

```python
def serach_pool():
    serach_frm = Frame(scr, highlightbackground="black" , highlightthickness=3)
    serach_frm.place(x=400, y=250, relwidth =.45, relheight=.45)

    lbl_user = Label(serach_frm,text="Welcome, User", font= ('',12,"bold"),fg="green")
    lbl_user.place(relx=0,rely=0)
    
    def back_to_main():
        serach_frm.destroy()
        index_frame()
    
    btn_back = Button(serach_frm, text="Back",font=('',12,'bold'),bd=3, command=back_to_main)
    btn_back.place(relx=.91,rely=.0)
```

```python
def new_pool():
    newpool_frm = Frame(scr, highlightbackground="black" , highlightthickness=3)
    newpool_frm.place(x=400, y=250, relwidth =.45, relheight=.45)

    lbl_user = Label(newpool_frm,text="Welcome, User", font= ('',12,"bold"),fg="green")
    lbl_user.place(relx=0,rely=0)
    
    def back_to_main():
        newpool_frm.destroy()
        index_frame()
    
    btn_back = Button(newpool_frm, text="Back",font=('',12,'bold'),bd=3, command=back_to_main)
    btn_back.place(relx=.91,rely=.0)
```
