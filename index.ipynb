```python
from tkinter.ttk import *
from tkinter import *
from tkinter import messagebox
from PIL import Image, ImageTk
import sqlite3
```

```python
try:
  con = sqlite3.connect(database='carpool.sqlite')
  cursor = con.cursor()
  table1 = 'CREATE TABLE users ("id"	INTEGER,"username"	text,	"password"	text,	"email"	text,	"number"	text,PRIMARY KEY ("id" AUTOINCREMENT))'
  table2 = 'CREATE TABLE cars("carno" text primary key, "model" text,"src" text, "dest" text, "seat" text, "fuel" type)'
  cursor.execute(table1)
  cursor.execute(table2)
  con.commit()
  con.close()
  print("table created")
except Exception as e:
  print('An exception occurred')
```

```python
# Screen
scr = Tk()

# Main Frame 
def main_frame():
    
    # Creating a full screen view
    scr.state("zoomed")
    scr.configure(bg="powder blue")
    
    # Site Title
    lbl_title = Label(scr, text= "Car Pooling", font=('',80,"bold","underline"), bg= "powder blue")
    lbl_title.pack()

    # Images / Logos 
    logo = Image.open('img\logo.jpg').resize((400,150))
    logotk = ImageTk.PhotoImage(logo, master =scr )
    lbl_logo = Label(scr, image= logotk, highlightbackground="black",highlightthickness=3)
    lbl_logo.place(relx=0,rely=0)

    logo2 = Image.open('img\carpool.jpg').resize((400,150))
    logotk2 = ImageTk.PhotoImage(logo2, master = scr )
    lbl_logo2 = Label(scr, image= logotk2, highlightbackground="black",highlightthickness=3)
    lbl_logo2.place(relx=0.76,rely=0)

    scr.mainloop()
```

```python
# Login frame
def login_frame():
    
    # Adding a login frame
    login_frm = Frame(scr, highlightbackground="brown" , highlightthickness=3)
    login_frm.place(x=450, y=250, relwidth =.4, relheight=.45)
    
    def reset_login(event):
        e_user.delete(0,'end')
        e_pass.delete(0,'end')
        e_user.focus()
    
    def login_check(event):
        u = e_user.get()
        p = e_pass.get()
        try:
          con = sqlite3.connect('carpool.sqlite')
          cursor = con.cursor()
          cursor.execute("SELECT * FROM users WHERE username = ? AND password = ? ",(u,p))
          loginuser_row = cursor.fetchone()
          con.close()
          if loginuser_row == None:
            messagebox.showerror('Login', 'Invalid username and password')
          else:
            pass
        except:
          print('An exception occurred')
        
    
    # Adding Fields Username and Password
    lbl_user = Label(login_frm, text="Username:",font=('',18,"bold"))
    lbl_pass = Label(login_frm, text="Password:",font=('',18,"bold"))
    e_user = Entry(login_frm, font=('',18,"bold"), bd=3 )
    e_user.focus()
    e_pass = Entry(login_frm, font=('',18,"bold"), bd=3, show = "*" )
    lbl_user.place(x=60,y=50)
    e_user.place(x=240,y=50)
    lbl_pass.place(x=60,y=110)
    e_pass.place(x=240,y=110)
    
    # Funtion for login user
    def login():
        login_frm.destroy()
        index_frame()
        
    # Adding Button Login, Reset
    btn_login = Button(login_frm, text ="Login",font=('',16,"bold"), bd=3, command=login, width=12)
    btn_login.place(x=100,y=180)
    btn_login.bind('<Button>', login_check)
    
    btn_reset = Button(login_frm, text ="Reset",font=('',16,"bold"),bd=3,width=12)
    btn_reset.place(x=320,y=180)
    btn_reset.bind('<Button>', reset_login)
   
    # Funtion for calling New user Frame and Destroying Previous frame
    def newuser():
        login_frm.destroy()
        newuser_frame()
    
    # Add New user button 
    btn_new_user = Button(login_frm, text ="Create a New Account",font=('',16,"bold"), bd=3, command= newuser, width=30)
    btn_new_user.place(x=90,y=250)
```

```python
# New User Frame    
def newuser_frame():

    newuser_frm = Frame(scr, highlightbackground="brown" , highlightthickness=3)
    newuser_frm.place(x=400, y=250, relwidth =.5, relheight=.5)

    lbl_user = Label(newuser_frm,text="Welcome, User", font= ('',12,"bold"),fg="green")
    lbl_user.place(relx=0,rely=0)

    # Funtion for return to Login Frame
    def back_home():
        newuser_frm.destroy()
        login_frame()
    
    btn_logout = Button(newuser_frm,text="Back",font=('',12,"bold"),bd=3, command=back_home)
    btn_logout.place(relx=.925,rely=0)
    
    
    def reset_button(event):
        e_username.delete(0,'end')
        e_password.delete(0,'end')
        e_email.delete(0,'end')
        e_number.delete(0,'end')
        e_username.focus()
    
    # Registeration New user in Sqlite DB using Command for Insert  
    def reg_user(event):
        user = e_username.get()
        password = e_password.get()
        mobile = e_number.get()
        email = e_email.get()
        
        if(len(user) == 0 or len(password) == 0 or len(mobile) == 0 or len(email) == 0):
            messagebox.showwarning('Validation', 'Please fill all fields')
        else:
            try:
                con = sqlite3.connect(database="carpool.sqlite")
                cursor = con.cursor()
                cursor.execute('INSERT INTO users(username, password,email,number) VALUES(?,?,?,?)', (user, password,email,mobile))
                con.commit()
                con.close()
                messagebox.showinfo('Registration','Registration has been done')
            except:
                messagebox.showerror('error',str(e))

        
    #Username Textfield    
    lbl_uname = Label(newuser_frm, text="Username:",font=('',20,"bold"))
    lbl_uname.place(x = 130,y = 50)
    e_username = Entry(newuser_frm,font=('',20,"bold"), bd=3)
    e_username.place(x=320,y=50)

    # Password Textfield
    lbl_pass = Label(newuser_frm, text="Password:",font=('',20,"bold"))
    lbl_pass.place(x=130,y=100)
    e_password = Entry(newuser_frm, font=('',20,'bold'), bd=3,show = "*")
    e_password.place(x=320, y=100)

    # Email textfield
    lbl_email = Label(newuser_frm, text="Email:",font=('',20,"bold"))
    lbl_email.place(x=130,y=150)
    e_email = Entry(newuser_frm, font=('',20,'bold'), bd=3)
    e_email.place(x=320, y=150)

    # Mobile Number textfield
    lbl_number = Label(newuser_frm, text="Mobile No.:",font=('',20,"bold"))
    lbl_number.place(x=130,y=200)
    e_number = Entry(newuser_frm, font=('',20,'bold'), bd=3)
    e_number.place(x=320, y=200)

    # Button for Adding New User
    btn_reg = Button(newuser_frm,text="Register",font=('',16,"bold"))
    btn_reg.place(x=250,y=300)
    btn_reg.bind('<Button>', reg_user)
    
    # Back Button 
    btn_reset = Button(newuser_frm,text="Reset",font=('',16,"bold"))
    btn_reset.place(x=450,y=300)
    btn_reset.bind('<Button>', reset_button)
```

```python
def index_frame():
    
    # Frame After login into site
    index_frm = Frame(scr, highlightbackground="brown" , highlightthickness=3)
    index_frm.place(x=400, y=250, relwidth =.45, relheight=.45)
  
    lbl_user = Label(index_frm,text="Welcome, User", font= ('',12,"bold"),fg="green")
    lbl_user.place(relx=0,rely=0)

    # Funtion for return to Login Frame
    def back_home():
        index_frm.destroy()
        login_frame()
  
    btn_logout = Button(index_frm,text="Logout",font=('',12,"bold"),bd=3, command=back_home)
    btn_logout.place(relx=.89,rely=0)
  
    # Funtion for jump to search frame
    def search():
        index_frm.destroy()
        serach_pool()
    
    # Funtion for jump to create new pool
    def newpool():
        index_frm.destroy()
        new_pool()
        
    def profile():
        index_frm.destroy()
        userprofile()
    
    # Button for Search pools
    btn_search = Button(index_frm,text="Search CarPool",font=('',16,"bold"),bd=3,width=15, command=search)
    btn_search.place(relx=.35,rely=.2)
    
    # Button for Create new pool
    btn_newpool = Button(index_frm,text="Create CarPool",font=('',16,"bold"),bd=3,width=15, command=newpool)
    btn_newpool.place(relx=.35,rely=.4)
    
    # Button for user profile
    btn_profile = Button(index_frm,text="User Profile",font=('',16,"bold"),bd=3,width=15, command=profile)
    btn_profile.place(relx=.35,rely=.6)
```
